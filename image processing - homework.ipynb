```python
# Load the os library
import os
import matplotlib.pyplot as plt
import numpy as np

files = [os.path.join('img_align_celeba', file_i)
 for file_i in os.listdir('img_align_celeba')
 if '.jpg' in file_i]

# There should be 100 files, with the last one being number 99
im = plt.imread(files[8])
plt.imshow(im)
```

## Grayscale conversion

```python
def to_grayscale(im, weights = np.c_[0.2989, 0.5870, 0.1140]):
    """
    Transforms a colour image to a greyscale image by
    taking the mean of the RGB values, weighted
    by the matrix weights
    """
    tile = np.tile(weights, reps=(im.shape[0],im.shape[1],1))
    return np.sum(tile * im, axis=2)

img_grey = to_grayscale(im)

fig, ax = plt.subplots(1, 2, sharey = True)
ax[0].imshow(img_grey, cmap = 'Greys')
ax[1].imshow(img_grey, cmap = 'gray')
```

# Colours

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for c, ax in zip(range(3), axs):
    tmp_im = np.zeros(im.shape, dtype="uint8")
    tmp_im[:,:,c] = im[:,:,c]
    ax.imshow(tmp_im)
    ax.set_axis_off()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(im[:,:,0], cmap='Reds')
ax[0].set_title('Red')
ax[1].imshow(im[:,:,1], cmap='Greens')
ax[1].set_title('Green')
ax[2].imshow(im[:,:,2], cmap='Blues')
ax[2].set_title('Blue');
```

# Negative of an image


```python
fig = plt.figure(figsize=(10, 10))

negative_grey = img_grey / (255*3)
fig.add_subplot(1, 2, 1)
plt.imshow(negative_grey)
plt.title('Negative of Grey image')

negative_rgb = 255 - im
fig.add_subplot(1, 2, 2)
plt.imshow(negative_rgb)
plt.title('Negative of RGB image')
```

# Color Reduction

```python
# Making Pixel values discrete by first division by // which gives int and then multiply by same factor

img_0 = (im // 64) * 64    
img_1 = (im // 128) * 128

img_all = np.concatenate((img_0, img_1), axis=1)

plt.figure(figsize=(15, 15))
plt.imshow(img_all)
```
